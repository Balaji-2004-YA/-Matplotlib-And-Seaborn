```python
import seaborn as sns
import matplotlib.pylab as plt
```

```python
##pre installed dataset 
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.scatterplot(x="total_bill",y="tip",data=tips)
```

distribution plots (histograms,KDE - kernal density estimates)

```python
sns.histplot(tips['total_bill'],kde=True)
plt.title("distribution of total bills ")
plt.show()


##KDE plot of tips
sns.kdeplot(tips['tip'])
plt.title("density of tips")
plt.show()
```

```python
## violinplot
```

```python
sns.violinplot(x=["A","A","B","B"],y=[10,20,15,25])
plt.show()
```

```python
#heat map
```

```python
import numpy as np
```

```python
data = np.random.rand(4,4)
sns.heatmap(data,annot=True,cmap="coolwarm")
plt.show()
```

```python
##pair plot
```

```python
iris=sns.load_dataset("iris")
sns.pairplot(iris,hue="species")
plt.show()
```

```python
##join plot
```

```python
sns.jointplot(x="sepal_length", y='sepal_width', data=iris, kind="scatter")
plt.show()
```

```python
##regplot
```

```python
sns.regplot(x=[1,2,3,4],y=[10,20,30,45])
plt.show()
```

```python
###box plot
```

```python
sns.boxplot(x="day",y="total_bill",data=tips)
plt.title("box plot of total bill")
plt.show()
```

```python
##cat plot
```

```python
sns.catplot(x="species",y="sepal_length",data=iris,kind="box")
plt.show()
```

```python
sns.catplot(x="species",y="sepal_length",data=iris,kind="bar")
plt.show()
```
