```python
import numpy as np
```

```python
x = np.linspace(1,10,100)
```

```python
y=np.sin(x)
```

```python
plt.plot(x,y)
plt.title("sine wave")
plt.xlabel("sin(X)")
plt.ylabel("x values")
plt.show()
```

```python
plt.plot(x,y, color="green",linestyle="--",marker="o")
plt.title("customized sine wave",fontsize=14,fontweight="bold")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.grid(True)
plt.show()
```

Bar plot

```python
categories=['A','B','C','D','E']
values=[3,2,5,6,1]
```

```python
plt.bar(categories,values,color=['red','yellow','green','orange','purple'])
plt.title("catagory comparison",fontsize=10,fontweight="bold")
plt.xlabel("catagory",fontsize=10,fontweight="bold")
plt.ylabel("values",fontsize=10,fontweight="bold")
plt.show()
```

HISTOGRAMS

```python
data=np.random.randn(1000)
```

```python
plt.hist(data,bins=30,color="green",alpha=0.2)
plt.title("random data histograms",fontsize=10,fontweight="bold")
plt.xlabel("x values",fontsize=10,fontweight="bold")
plt.ylabel("y values",fontsize=10,fontweight="bold")
plt.show()
```
